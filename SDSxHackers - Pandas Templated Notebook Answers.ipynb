## Applying Functions

Sorry! I cannot intepret what Survived = 1, and Survived = 0 means. Can you change it into something more meaningful so that I can understand? Store this new dataset into a new variable df2.

```python
def has_survived(x): 
  if x == 1:
    return 'Survived' #returns Survived if 1
  else:
    return 'Died' #returns Died if 0

df2 = df.copy() #Creates a copy of df and storing it as df2
df2["Survived"] = df2["Survived"].apply(has_survived) #apply function on column
df2 #show df2 again
```
