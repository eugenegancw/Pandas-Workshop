## Merge Function

```python
df_merge1 = df9[['PassengerId', 'Name']]  #Before
df_merge1
```

```python
df_merge2 = df9[['PassengerId','Ticket','Age']] #Before
df_merge2 
```

```python
df_merge1.merge(df_merge2, on= 'PassengerId') #by default, innner join
```

Perform ***Left*** Join via merge function



```python
df_merge1.merge(df_merge2, on= 'PassengerId', how='left') 
```

```python
df_merge1.merge(df_merge2, left_on= 'PassengerId', right_on= 'PassengerId', how = 'left')
```

Perform ***Right*** Join via merge function


```python
df_merge1.merge(df_merge2, on= 'PassengerId', how='right')
```

```python
df_merge1.merge(df_merge2, left_on= 'PassengerId', right_on= 'PassengerId', how = 'right')
```

Perform ***Outer*** Join via merge function


```python
df_merge1.merge(df_merge2, on= 'PassengerId', how='outer')
```

```python
df_merge1.merge(df_merge2, left_on= 'PassengerId', right_on= 'PassengerId', how = 'outer')
```
